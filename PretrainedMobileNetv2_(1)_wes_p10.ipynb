Building Model


```python
from keras.layers import Conv2D, AveragePooling2D, Dropout, Flatten, Dense
from keras.optimizers import RMSprop
```

```python
my_model = Sequential()
base_model = MobileNetV2(weights = 'imagenet',include_top=False,input_shape=(224,224,3), classes=3)
for layer in base_model.layers:
        layer.trainable=False
my_model.add(base_model)
my_model.add(Conv2D(32, (3, 3), padding='same', activation='relu'))
my_model.add(AveragePooling2D(pool_size=(2, 2)))

# Add the dropout layer
my_model.add(Dropout(0.5))

# Add the flatten layer
my_model.add(Flatten())

# Add the dense layer
my_model.add(Dense(3, activation='softmax'))

# Compile the model
optimizer = RMSprop(learning_rate=0.001)
my_model.compile(optimizer=optimizer, loss='categorical_crossentropy', metrics=['accuracy'])
my_model.summary()


```

```python
# epochs=20

# history = my_model.fit(
#     train_ds,
#     validation_data=val_ds,
#     epochs=epochs,
# )


import tensorflow as tf

# ... (your model definition and data preparation)

epochs = 300

# Early stopping callback with best weight restoration
early_stopping = tf.keras.callbacks.EarlyStopping(
    monitor='val_loss',  # Monitor validation loss
    patience=10,        # Stop training if validation loss doesn't improve for 10 epochs
    restore_best_weights=True  # Restore weights from the best epoch
)

# Model checkpoint callback (for saving best weights)
model_checkpoint = tf.keras.callbacks.ModelCheckpoint(
    filepath='/content/drive/MyDrive/checkpoints/MobileNetv2/best_model_wes_p10.h5',  # Save best model in Drive's checkpoints folder
    save_best_only=True,       # Only save the model with the best validation loss
    monitor='val_loss',        # Monitor validation loss for saving
    mode='min'                 # Save model with the lowest validation loss
)

# Check for existing checkpoint (using tf.train.CheckpointManager)
checkpoint_dir = '/content/drive/MyDrive/checkpoints/MobileNetv2'  # Adjust checkpoint directory if needed
if not hasattr(tf.train.CheckpointManager, 'checkpoint_dir'):
    # For older TensorFlow versions
    checkpoint = tf.train.CheckpointManager(my_model, directory=checkpoint_dir, max_to_keep=5)
else:
    # For newer TensorFlow versions
    checkpoint = tf.train.CheckpointManager(my_model, checkpoint_dir=checkpoint_dir, max_to_keep=5)

# Optionally load weights from the latest checkpoint
if checkpoint.latest_checkpoint:
    checkpoint.restore(checkpoint.latest_checkpoint)
    print('Loaded weights from latest checkpoint.')

# Train the model with callbacks
history = my_model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=epochs,
    callbacks=[early_stopping, model_checkpoint]
)
```

```python
test_loss, test_accuracy = my_model.evaluate(test_ds)
print("Test Loss:", test_loss)
print("Test Accuracy:", test_accuracy)
```

```python
fig1 = plt.gcf()
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.axis(ymin=0.4,ymax=1)
plt.grid()
plt.title('Model Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epochs')
plt.legend(['train', 'validation'])
plt.show()
```
